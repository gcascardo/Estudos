```python
temperatura_por_dia_semana = df.groupby('dia_da_semana')['temperatura'].mean()
dia_da_semana = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_semana = temperatura_por_dia_semana[dia_da_semana]
temperatura_por_dia_semana
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_semana))

eixo.bar(indice, temperatura_por_dia_semana)
eixo.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
eixo.set_xlabel('Dia da Semana', fontsize = 15)
eixo.set_ylabel('Temperatura média', fontsize = 15)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_semana))

eixo.bar(indice, temperatura_por_dia_semana)
eixo.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
eixo.set_xlabel('Dia da Semana', fontsize = 15)
eixo.set_ylabel('Temperatura média', fontsize = 15)
eixo.set_xticks(indice)
eixo.set_xticklabels(dia_da_semana)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'gray', 'magenta']

eixo.bar(indice, temperatura_por_dia_semana, color = cores)
eixo.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
eixo.set_xlabel('Dia da Semana', fontsize = 15)
eixo.set_ylabel('Temperatura média', fontsize = 15)
eixo.set_xticks(indice)
eixo.set_xticklabels(dia_da_semana)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'gray', 'magenta']

eixo.bar(indice, temperatura_por_dia_semana, color = cores, edgecolor = 'black')
eixo.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
eixo.set_xlabel('Dia da Semana', fontsize = 15)
eixo.set_ylabel('Temperatura média', fontsize = 15)
eixo.set_xticks(indice)
eixo.set_xticklabels(dia_da_semana)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_semana, labels = temperatura_por_dia_semana.index)
eixo.set_title('Temperatura por dia da Semana', fontsize = 20, pad = 10)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_semana, labels = temperatura_por_dia_semana.index,
        autopct = '%.1f%%')
eixo.set_title('Temperatura por dia da Semana', fontsize = 20, pad = 10)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.3, 0, 0, 0, 0, 0.3, 0.3]

eixo.pie(temperatura_por_dia_semana, labels = temperatura_por_dia_semana.index,
        autopct = '%.1f%%', explode = explodir)
eixo.set_title('Temperatura por dia da Semana', fontsize = 20, pad = 10)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.3, 0, 0, 0, 0, 0.3, 0.3]

eixo.pie(temperatura_por_dia_semana, labels = temperatura_por_dia_semana.index,
        autopct = '%.1f%%', explode = explodir, shadow = True)
eixo.set_title('Temperatura por dia da Semana', fontsize = 20, pad = 10)
```
